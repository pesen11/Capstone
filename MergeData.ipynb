```python
#Firstly lets deal with the homicide dataset
```

```python
# the homicide type has shooting, stabbing and others. replace it with offence so we can merge with offence in the crimeData df.
homicide.rename(columns={"HOMICIDE_TYPE":"OFFENCE"},inplace=True)
```

```python
homicide.head()
```

```python
# The shooting dataframe has no offence column. so let us create a offence column and add Shooting for all of its values
shooting["OFFENCE"]="Shooting"
```

```python
homicide["MCI_CATEGORY"]="Homicide"
```

```python
#for some shooting cases, they may be a homicidal case, so check with the homicide data. For those shooting cases assign category homicide
#check the event_unique_id of shooting with that of homicide, if they match, assign 'Homicide' for those shooting records
```

```python
homicide_event_ids = set(homicide['EVENT_UNIQUE_ID'])

# Assign "Homicide" where EVENT_UNIQUE_ID is in the homicide DataFrame, else assign NaN
shooting['MCI_CATEGORY'] = shooting['EVENT_UNIQUE_ID'].apply(
    lambda x: 'Homicide' if x in homicide_event_ids else pd.NA
)
```

```python
#also the shooting records may also be present in crimeData. if they are just assign the mci_category from crimeData
```

```python

# Perform a left join between shooting_df and crimeData on EVENT_UNIQUE_ID
# Bring in the 'MCI_CATEGORY' from crimeData for matching records
merged_df = pd.merge(shooting, crimeData[['EVENT_UNIQUE_ID', 'MCI_CATEGORY']], 
                     on='EVENT_UNIQUE_ID', how='left', suffixes=('', '_from_crimeData'))

# Update 'MCI_CATEGORY' in shooting_df only where the value from crimeData is not NaN
merged_df['MCI_CATEGORY'] = merged_df['MCI_CATEGORY_from_crimeData'].combine_first(merged_df['MCI_CATEGORY'])

# Drop the extra column created from crimeData
merged_df.drop(columns=['MCI_CATEGORY_from_crimeData'], inplace=True)


merged_df
```

```python
shooting_categorized=merged_df
```

```python
# also for homicide dataframe we do not have occ_hour column. lets get that from shooting
```

```python
homicide["OCC_HOUR"]=np.nan
```

```python
homicide_occHr = pd.merge(homicide, shooting_categorized[['EVENT_UNIQUE_ID', 'OCC_HOUR']], 
                           on='EVENT_UNIQUE_ID', how='left', suffixes=('', '_from_shooting'))

# Update the 'OCC_HOUR' in homicide_df with the value from shooting_df where available
homicide_occHr['OCC_HOUR'] = homicide_occHr['OCC_HOUR_from_shooting'].combine_first(homicide_occHr['OCC_HOUR'])

# Drop the extra column created from shooting_df
homicide_occHr.drop(columns=['OCC_HOUR_from_shooting'], inplace=True)
```

```python
homicide_categorized=homicide_occHr
```

```python

# Keep relevant columns
homicide_df = homicide_categorized[['OBJECTID', 'EVENT_UNIQUE_ID', 'OCC_DATE', 'OCC_YEAR', 'OCC_MONTH', 
                           'OCC_DAY', 'OCC_DOW', 'OCC_DOY', 'OFFENCE','MCI_CATEGORY','OCC_HOUR',
                           'LONG_WGS84', 'LAT_WGS84', 'HOOD_158', 'NEIGHBOURHOOD_158']]

shooting_df = shooting_categorized[['OBJECTID', 'EVENT_UNIQUE_ID', 'OCC_DATE', 'OCC_YEAR', 'OCC_MONTH', 
                           'OCC_DAY', 'OCC_DOW', 'OCC_DOY','OCC_HOUR', 'OFFENCE','MCI_CATEGORY',
                           'LONG_WGS84', 'LAT_WGS84', 'HOOD_158', 'NEIGHBOURHOOD_158']]

crimeData_df = crimeData[['OBJECTID', 'EVENT_UNIQUE_ID', 'OCC_DATE', 'OCC_YEAR','OCC_MONTH','OCC_DAY','OCC_DOW','OCC_DOY','OCC_HOUR',
                              'OFFENCE', 'MCI_CATEGORY', 'LONG_WGS84', 'LAT_WGS84', 
                              'HOOD_158', 'NEIGHBOURHOOD_158']]

```

```python
# Concatenate the DataFrames
concatenated_df = pd.concat([homicide_df, shooting_df], ignore_index=True)

# Remove duplicates based on 'event_unique_id', keeping the first occurrence
final_homicide_shooting = concatenated_df.drop_duplicates(subset='EVENT_UNIQUE_ID', keep='first')
```

```python
final_homicide_shooting
```

```python
# Concatenate the DataFrames
final_merged = pd.concat([final_homicide_shooting, crimeData_df], ignore_index=True)

# Remove duplicates based on 'event_unique_id', keeping the first occurrence
final_merged = final_merged.drop_duplicates(subset='EVENT_UNIQUE_ID', keep='first')
```

```python
merged_df=final_merged
```

```python
def categorize_time(hour):
    if pd.isna(hour):
        return np.nan
    elif 21 <= hour < 24 or 0 <= hour < 4:
        return 'Night'
    elif 4 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    else:
        return 'Evening'
```

```python
merged_df['OCC_TIME_RANGE'] = merged_df['OCC_HOUR'].apply(categorize_time)
```
